```python
import re
import os
```

```python
# 시그니처 불러오기

def load_signature():
    signature = {}
    
    f = open('Signature.txt')
    lines = f.readlines()
    for line in lines:
        row = line.strip().split(',')
        signature[row[0]] = row[1]
    
    return signature

signature = load_signature()

load_signature()
```

```python
# Tokenizer 작성
# https://stackoverflow.com/questions/49950961/how-to-tokenize-source-code-of-other-programming-language

def tokenize(data):
    syntax = [
        r'/\*.*?\*/',
        r'".*?"',
        r'\'.*?\'',
        r'\$[\w\d_]+', # variable name
        r'\w+', # function name
        r'return',
        r'<\?php',
        r'=>',
        r'\?>',
        r'\[',
        r'\]',
        r',',
        r';',
        r'@',
        r'\(',
        r'\)',
        r'\.',
        r'\n',
        r'\s',
        r'=',
        r'\W',
    ]

    s = r'(' + r'|'.join(syntax) + r')'
    print(s)

    #tokens = re.split(s, self._source, re.DOTALL | re.M | re.I | re.UNICODE)
    tokens = re.findall(s, data)

    return tokens
```

```python
# 파일 처리
# https://stackoverflow.com/questions/27453879/unicode-decode-error-how-to-skip-invalid-characters

def load_file(filename):
    testdata = b''
    with open(filename, 'rb') as f:
        testdata = f.read()
    
    data = testdata.decode('utf-8', errors='ignore')

    return data

print('\r\n > '.join(tokenize(load_file('./Data/1ea24b3ae95b983c7a471a684bc980fb.vt'))))
```

```python
# 토큰 처리
def processing_tokens(tokens):
    detected_signature = {}
    
    for token in tokens:
        if token in signature:
            signature_category = signature[token]
            
            if signature_category in detected_signature:
                detected_signature[signature[token]] += 1
            else:
                 detected_signature[signature[token]] = 1

    return detected_signature

# 파일 불러오기
data_directory = 'Data'
for root, dirs, files in os.walk(data_directory):
    for file in files:
        detected_signature = {}
        
        filepath = os.path.join(data_directory, file)
        contents = load_file(filepath)
        tokens = tokenize(contents)
        detected_signature = processing_tokens(tokens)

        print (filepath)
        print (contents[0:100])
        for k in detected_signature:
            print(k, detected_signature[k])
        print ()
```
